```python
# Import Numpy library
import numpy as np
```

# Numpy Array Creation

```python
list1 = [10,20,30,40,50,60]
list1
```

```python
# Display the type of an object
type(list1)
```

```python
# Convert List to Numpy Array
arr1 = np.array(list1)
arr1
```

```python
# Memory address of an array object
arr1.data
```

```python
# Display type of an object
type(arr1)
```

```python
# Datatype of array
arr1.dtype
```

```python
# Convert Integer Array to Float
arr1.astype(float)
```

```python
# Generate evenly spaced numbers (space = 1) between 0 to 10
np.arange(0,10)
```

```python
# Genrate numbers between 0 to 100 with a space of 10
np.arange(0,100,10)
```

```python
# Genrate numbers between 10 to 100 with a space of 10 descending order
np.arange(100,10,-10)
```

```python
# Shape of array
arr3 = np.arange(0,10)
print(arr3.shape)
print(arr3)
```

```python
# Size of array
arr3.size
```

```python
# Dimension
arr3.ndim
```

```python
# Datatype of object
arr3.dtype
```

```python
# Bytes comsumed by one element of an array object
arr3.itemsize
```

```python
# Length of array
len(arr3)
```

```python
# Generate an array of zeros
np.zeros(10)
```

```python
# Generate an array of ones with given shape
np.ones(10)
```

```python
# Repeat 10 five times in an array
np.repeat(10,5)
```

```python
# Repeat each element in array 'a' thrice
a = np.array([10,20,30])
np.repeat(a,3)
```

```python
# Array of 10's
np.full(5,10)
```

```python
# Generate array of Odd numbers
ar1 = np.arange(1,20)
ar1[ar1%2 == 1]
```

```python
# Generate array of even numbers

ar1 = np.arange(1,20)
ar1[ar1%2 == 0]
```

```python
# Generate evenly spaced 4 numbers between 10 to 20
np.linspace(10,20,4)
```

```python
# Create an array of random values
np.random.random(4)
```

```python
# Generate an array of Random Integer numbers
print(np.random.randint(0, 500, 5))
print(np.random.randint(0, 500, 10))
```

```python
# Using random.seed we can generate same number of Random numbers

np.random.seed(123)
np.random.randint(0, 100, 10)
```

```python
# Generate array of Random float numbers
f1 = np.random.uniform(5,10, size=(10))
f1
```

```python
# Extract Integer part
np.floor(f1)
```

```python
# Truncate decimal part
np.trunc(f1)
```

```python
# Convert Float Array to Integer array
f1.astype(int)
```

```python
# Normal distribution (mean=0 and variance=1)

b2 = np.random.randn(10)
b2
```
