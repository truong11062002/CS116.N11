```python
import geopandas as gpd
import numpy as np
import rtree
import matplotlib.pyplot as plt
import folium
import osmnet
import geojson
import pygeos
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Colab Notebooks/cs116_python_ml/CSL_HCMC/Data/GIS/Population/population_HCMC/population_shapefile/Population_District_Level.shp'
```

```python
data = gpd.read_file(path)
data.head()
```

```python
data.plot(figsize=(6, 6))
plt.show()
```

```python
data.plot(column=data['Dist_Name'], cmap='winter', alpha=2, categorical=None, legend=None, axes=None,)
```

```python
data.crs
```

```python
data = data.to_crs(epsg=4326)
print(data.crs)
data.head()
```

```python
data.plot(figsize=(6, 6))
plt.show()
```

```python
data.plot(column=data['Dist_Name'], cmap='winter', alpha=2, categorical=None, legend=None, axes=None,)
```

```python
m = folium.Map(zoom_start= 10, tiles='CartoDB positron')
m
```

```python
color = ['red', 'blue', 'green']

import random
```

```python
for _, r in data.iterrows():
    # Without simplifying the representation of each borough,
    # the map might not be displayed
    sim_geo = gpd.GeoSeries(r['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j,
                           style_function=lambda x: {'fillColor': 'orange'})
    folium.Popup(r['Dist_Name']).add_to(geo_j)
    geo_j.add_to(m)
```

```python
for _, r in data.iterrows():
    # Without simplifying the representation of each borough,
    # the map might not be displayed
    sim_geo = gpd.GeoSeries(r['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j,
                           style_function=lambda x: {'fillColor': color[random.randint(0,2)]})
    folium.Popup(r['Dist_Name']).add_to(geo_j)
    geo_j.add_to(m)
```

```python
data = data.to_crs(epsg=2263)

# Access the centroid attribute of each polygon
data['centroid'] = data.centroid
```

```python
data = data.to_crs(epsg=4326)

data['centroid'] = data['centroid'].to_crs(epsg=4326)

data.head()
```

```python
for _, r in data.iterrows():
    lat = r['centroid'].y
    lon = r['centroid'].x
    folium.Marker(location=[lat, lon],
                  popup='length: {} <br> area: {}'.format(r['Shape_Leng'], r['Shape_Area'])).add_to(m)
```
